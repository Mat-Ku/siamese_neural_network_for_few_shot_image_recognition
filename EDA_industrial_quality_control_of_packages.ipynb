## Exploratory Data Analysis of Package Images

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
data_path = '*'  # for * insert path to dataset
```

```python
# Gather file names of all images in nested dict
images = {'damaged': {'side': [],
                     'top': []},
         'intact': {'side': [],
                   'top': []}
         }

for path, directories, files in os.walk(data_path):
    for file in files:
        f = os.path.join(path, file).replace('\\', '/')
        if '/damaged' in f:
            if '/side' in f:
                images['damaged']['side'].append(f)
            else:
                images['damaged']['top'].append(f)
        else:
            if '/side' in f:
                images['intact']['side'].append(f)
            else:
                images['intact']['top'].append(f)
```

```python
print(f"# of pictures of damaged packages taken from the side: {len(images['damaged']['side'])}")
print(f"# of pictures of damaged packages taken from top: {len(images['damaged']['top'])}")
print(f"# of pictures of intact packages taken from the side: {len(images['damaged']['side'])}")
print(f"# of pictures of intact packages taken from top: {len(images['damaged']['top'])}")
```

```python
# Show first 5 damaged packages
for i in range(0, 5):
    img_s = mpimg.imread(images['damaged']['side'][i])
    plt.imshow(img_s)
    plt.axis('off')
    plt.show()
    
    img_t = mpimg.imread(images['damaged']['top'][i])
    plt.imshow(img_t)
    plt.axis('off')
    plt.show()
```

```python
# Show first 5 intact packages
for i in range(0, 5):
    img_s = mpimg.imread(images['intact']['side'][i])
    plt.imshow(img_s)
    plt.axis('off')
    plt.show()
    
    img_t = mpimg.imread(images['intact']['top'][i])
    plt.imshow(img_t)
    plt.axis('off')
    plt.show()
```
